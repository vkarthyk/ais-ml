```python
def getAllFieldsAndAttr(dictOfFields,itor):
    stixfield=[]
    try:
        while True:
            stixfield.append(itor.next())#load all field in stix to a list
            #print len(stixfield)
    except StopIteration:
        pass
    
    for field in stixfield:
        name_of_attrs=[]
        type_and_name = field.typed_fields_with_attrnames()
        for (name,attr_obj) in type_and_name:
            if name not in name_of_attrs:
        #print str(type_name)
                name_of_attrs.append(name)
            else:
                pass
        try:
            if len(name_of_attrs)>len(dictOfFields[str(type(field))]):
                dictOfFields[str(type(field))]=name_of_attrs
                #print str(type(field))+'='+str(name_of_attrs)
            else:
                pass
        except KeyError:
            dictOfFields[str(type(field))]=name_of_attrs
            #print str(type(field))+'='+str(name_of_attrs)
```

```python
import os
comperhensive_dict={}
# for all the stix file
#from stix.core import STIXPackage        # Import the STIX Package API
#list_of_files = {}
path='.'
for (dirpath, dirnames, filenames) in os.walk(path):
    for filename in filenames:
        if filename.startswith('stix'):
            stix=STIXPackage.from_xml(filename)  # Parse using the from_xml() method
            getAllFieldsAndAttr(comperhensive_dict,stix.walk())

```
